```python
import pandas as pd

schema_play = ['uid','device','song_id','song_type','song_name','singer','play_time','song_length','paid_flag','file_name']
df = pd.read_csv('../music_box_data/data/play/20170303_1_play.log.fn',delimiter='\t',header=None,index_col=None,names=schema_play)
```

```python
df.iloc[1].str.split(',')
```

```python
df.head(10)
```

```python
# add new feature - date
def get_date(file_name):
    tmp_list = str(file_name).split('_')
    return tmp_list[0]
df['date'] = df['file_name'].map(get_date)
```

```python
df.iloc[350467]
```

```python
df.dtypes
```

```python
# convert play_time & song_length data type from object to float, invalid parsing will be set as NaN
df_play = df.copy()
df_play['play_time'] = pd.to_numeric(df_play['play_time'], errors='coerce')
df_play['song_length'] = pd.to_numeric(df_play['song_length'], errors='coerce')
```

```python
# problem here: song_length cannot transfer to float
df_play.dtypes
```

```python
# statistics of play_time
df_play.play_time.mean(),df_play.play_time.max(),df_play.play_time.min(),df_play.play_time.median()
```

```python
# get the row number of max in column play_time
df_play.play_time.idxmax(1)
```

```python
# how many rows are over 86400 second (24 hours) in play_time
df_play.loc[df_play['play_time'] > 86400]  #30929 rows
```

```python
# how many rows are less than 0 second in play_time
df_play.loc[df_play['play_time'] < 0]  
```

```python
# statistics of song_length
df_play.song_length.mean(),df_play.song_length.max(),df_play.song_length.min(),df_play.song_length.median()
```

```python
# how many rows are equals 0 second in song_length
df_play.loc[df_play['song_length'] == 0] # 223210 rows
```

```python
# how many rows are less than 0 second in song_length
df_play.loc[df_play['song_length'] < 0] # 545 rows
```

```python
# how many rows are more than 1800 second (30 min) in song_length
df_play.loc[df_play['song_length'] > 1800] # 10717 rows
```
